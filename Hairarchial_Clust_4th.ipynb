```python
# 📌 Importing Required Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.preprocessing import StandardScaler

# 📊 Step 1: Create Sample Data (Age, Annual Income)
data = np.array([
    [25, 15000],
    [28, 16000],
    [30, 18000],
    [35, 22000],
    [40, 25000],
    [45, 60000],
    [50, 65000],
    [55, 70000]
])

# 🧼 Step 2: Standardize the Data
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# 🌿 Step 3: Apply Hierarchical Clustering (Ward linkage)
model = AgglomerativeClustering(n_clusters=3, linkage='ward')
cluster_labels = model.fit_predict(data_scaled)

# 🖨️ Output cluster labels
print("Cluster Labels:", cluster_labels)

# 🖼️ Step 4: Visualize Clusters
plt.figure(figsize=(8, 5))
plt.scatter(data[:, 0], data[:, 1], c=cluster_labels, cmap='viridis', s=100)
plt.title("Customer Segmentation using Hierarchical Clustering")
plt.xlabel("Age")
plt.ylabel("Annual Income")
plt.grid(True)
plt.show()

# 🌿 Step 5: Plot Dendrogram
linked = linkage(data_scaled, method='ward')

plt.figure(figsize=(10, 6))
dendrogram(linked,
           orientation='top',
           distance_sort='descending',
           show_leaf_counts=True)
plt.title("Dendrogram (Ward Linkage)")
plt.xlabel("Sample Index")
plt.ylabel("Distance")
plt.show()
```

```python
# 📦 Import Required Libraries
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster
from sklearn.datasets import make_blobs

# 🎯 Step 1: Generate Sample Data
# Generates 10 data points around 3 centers
X, _ = make_blobs(n_samples=10, centers=3, random_state=42)

# 🌿 Step 2: Compute Linkage Matrix Using Ward's Method
linkage_matrix = linkage(X, method='ward')

# 🌲 Step 3: Plot Dendrogram
plt.figure(figsize=(12, 8))
dendrogram(
    linkage_matrix,
    labels=['P1', 'P2', 'P3', 'P4', 'P5', 'P6', 'P7', 'P8', 'P9', 'P10'],  # Custom labels for leaf nodes
    leaf_rotation=90,
    leaf_font_size=12
)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Data Points')
plt.ylabel('Distance')
plt.grid(True)
plt.show()

# 🔍 Step 4: Determine Cluster Assignments (e.g., 3 clusters)
clusters = fcluster(linkage_matrix, t=3, criterion='maxclust')

# 🖨️ Output Cluster Assignments
print("Cluster assignments:", clusters)
```

```python
# 📦 Import Required Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import StandardScaler
from scipy.cluster.hierarchy import dendrogram, linkage, fcluster

# 🎯 Step 1: Generate Sample Customer Data
np.random.seed(42)
n_customers = 100

data = pd.DataFrame({
    'Age': np.random.normal(40, 15, n_customers),
    'Income': np.random.normal(50000, 20000, n_customers),
    'SpendingScore': np.random.normal(50, 25, n_customers)
})

# 🧹 Step 2: Clean the Data
data = data[
    (data['Age'] > 0) &
    (data['Income'] > 0) &
    (data['SpendingScore'] >= 0) &
    (data['SpendingScore'] <= 100)
]

print("✅ Dataset shape:", data.shape)
print(data.head())

# 🧼 Step 3: Standardize the Features
scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)

# 🌿 Step 4: Compute Linkage Matrix (Ward Method)
linkage_matrix = linkage(data_scaled, method='ward')

# 🌲 Step 5: Plot Dendrogram
plt.figure(figsize=(15, 7))
dendrogram(
    linkage_matrix,
    truncate_mode='level',  # only show the top levels
    p=5                     # number of levels to show
)
plt.title('📊 Customer Segmentation Dendrogram')
plt.xlabel('Data Points or Clusters')
plt.ylabel('Distance')
plt.grid(True)
plt.show()
```

```python
# Cut the dendrogram to form clusters (say, 4 clusters)
cluster_labels = fcluster(linkage_matrix, t=4, criterion='maxclust')

# Add cluster labels to original data
data['Cluster'] = cluster_labels

# Pairplot or 2D plot to visualize clusters
sns.pairplot(data, hue='Cluster', palette='tab10')
plt.suptitle("Customer Segments", y=1.02)
plt.show()
```
